```python
import pandas as pd

kinetics_700_csv_path = "/checkpoint/abardes/datasets/Kinetics700/annotations/k700_train_paths.csv"
kinetics_400_csv_path = "/checkpoint/abardes/datasets/Kinetics400/annotations/k400_train_paths.csv"


df_400 = pd.read_csv(kinetics_400_csv_path, header=None, names=["video_path", "index"], sep=" ")
df_700 = pd.read_csv(kinetics_700_csv_path, header=None, names=["video_path", "index"], sep=" ")
```

```python
import os
import random
from IPython.display import Video, display

def show_random_video_from_df400(width=400, height=300, seed=None, check_exists=True):
    """
    Draws a random row from df_400, displays the corresponding video,
    and returns (row_index, video_path).

    Args:
        width, height: display size
        seed: optional int to make the random pick reproducible
        check_exists: if True, will resample until a path exists (up to a few tries)
    """
    if seed is not None:
        random.seed(seed)

    tries = 0
    while True:
        idx = random.randrange(len(df_400))
        video_path = df_400.at[idx, "video_path"]

        if not check_exists or os.path.exists(video_path) or tries >= 10:
            display(Video(video_path, embed=True, width=width, height=height))
            return idx, video_path

        tries += 1

# usage:
row_idx, path = show_random_video_from_df400()
print("Shown:", row_idx, path)
```
